# Remove outliers

```python
def remove_outliers(df , col , k=2):
        mean = df[col].mean()
        global df1
        sd = df[col].std()
        final_list = [x for x in df[col] if (x > mean - k * sd)]
        final_list = [x for x in final_list if (x < mean + k * sd)] 
        df1 = df.loc[df[col].isin(final_list)] ; print(df1.shape)
        print('Number of outlier removed == >' , df.shape[0] - df1.shape[0])
```

```python
remove_outliers(lcn,'LungCap' , k=2)
```

```python
remove_outliers(lcn,'LungCap' , k=3)
```

```python
remove_outliers(cr , "ApplicantIncome" , 3) # dont take value take less than 1
```

# join using pandas

```python
sales = pd.DataFrame({"PurchaseId": (111, 112,113,114 ,115 , 116, 117 , 118 ),
"ItemCode" :('phone', 'tv' , 'tablet' ,'phone' ,'pen' , 'food', 'cloth', 'shoes')})
```

```python
sales_dup = pd.DataFrame({"PurchaseId": (211, 212,213,114 ,115 , 116, 217 , 218 ),
"ItemCode" :('phone', 'tv' , 'tablet' ,'phone' ,'pen' , 'food', 'cloth', 'shoes')})
sales_dup
```

```python
df_inner = pd.merge(sales , sales_dup , on = "PurchaseId")
```

```python
df_inner
```

```python
df_left = pd.merge(sales , sales_dup , on = "PurchaseId" , how= "left")
df_left
```

```python
df_right = pd.merge(sales , sales_dup , on = "PurchaseId" , how= "right")
df_right
```
